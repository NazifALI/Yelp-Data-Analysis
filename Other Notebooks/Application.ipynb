```python
user_prob=pd.concat([raw_user['user_id'],raw_user['target'],raw_user['logstdScore'],raw_user['lognormScore'],raw_user['customProbEliteUser']],axis=1)
user_prob
```

```python
import collections
# ufalse=user_prob.loc[user_prob['target']==False]
# print user_prob['logstdScore'].mean()
dic_id_elite=collections.defaultdict()
for index,row in user_prob.iterrows():
    dic_id_elite[row['user_id']]=(row['target'],row['logstdScore'])
print "done"
print len(dic_id_elite)
print len(user_prob)
```

```python
import collections
freview=filteredreview

mean_score_deflation=user_prob['logstdScore'].mean()
dic_bus_reviews=collections.defaultdict()
# datestr=freview.loc[freview.index[0],'date']
# print type(freview.loc[freview.index[0],'stars'])
def date_to_tuples(s):
    s=s.split('-')
    return (int(s[0]),int(s[1]),int(s[2]))

# build dictionary of businsses to review with date and adjusted star
for index,row in filteredreview.iterrows():
#     test-=1
#     if test<=0: break
    bus=row['business_id']
    star=row['stars']
#   slow af
    user=user_prob.loc[user_prob['user_id']==row['user_id']]
    if len(user)==1:
        adjusted_star=dic_id_elite[row['user_id']][1]*star
    else:
        adjusted_star=star*mean_score_deflation
    if bus not in dic_bus_reviews:
        dic_bus_reviews[bus]=[(date_to_tuples(row['date']),star,adjusted_star)]
    else:
        dic_bus_reviews[bus].append((date_to_tuples(row['date']),star,adjusted_star))
print "done"
print len(dic_bus_reviews)
```

The following codes calculate the Time Series data to calculate 1. The overall trend of the restaurant 2. The average score of the restaurant and 3. The weighted average score of the restaurant, with the weight on the eliteness of users. The unit of slope is change in star rating per day.

```python
#dic_bus_review -- {business_id : (date,star,adjusted_star)}
#dic_id_elite -- {user_id: (is_elite,log_score)}
# for key,value in dic.iteritems():
regr = linear_model.LinearRegression()
d_old=date(2000,1,1)
# d_new=date(2012,8,3)
# k=(d_new-d_old).days

def int_to_date(d):
    return date(d[0],d[1],d[2])

def calc_slope(arr):
    x,y,adjusted_star=[],[],[]
    for t in arr:
        x.append((int_to_date(t[0])-d_old).days)
        y.append(t[1])
        adjusted_star.append(t[2])
    slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
    return slope,sum(y)/float(len(y)),sum(adjusted_star)/len(adjusted_star)

slope_arr,avg_arr,adj_arr=[],[],[]

for index,row in f_bus.iterrows():
    date_star_adj=dic_bus_reviews.get(row['business_id'])
    if date_star_adj is not None:
        slope,avg,adj=calc_slope(date_star_adj)
        f_bus.set_value(index,'slope',slope)
        f_bus.set_value(index,'average_stars',avg)
        f_bus.set_value(index,'average_adjusted_stars',adj)
        slope_arr.append(slope)
        avg_arr.append(avg)
        adj_arr.append(adj)
        

f_bus.head()
# def take(n, iterable):
#     return list(islice(iterable, n))

# n_items=take(100,dic_bus_reviews.iteritems())
# for key,value in n_items:
#     slope,avg,adj=calc_slope(value)
#     print key," ",slope," ",avg," ",adj

# ind=np.where(f_bus['business_id']=='UsFtqoBl7naz8AVUBZMjQQ')[0]
# f_bus.set_value(ind[0],'slope',?)
```
